```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
# In[2]:


iris = load_iris()
X = iris.data


```

```python
# In[3]:


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)



# In[ ]:
```

```python
# In[4]:


dbscan_hard = DBSCAN(eps=0.5, min_samples=5)
y_hard = dbscan_hard.fit_predict(X)


```

```python
# In[5]:


dbscan_hard = DBSCAN(eps=0.5, min_samples=5)
y_hardpca = dbscan_hard.fit_predict(X_pca)



```

```python
# In[6]:


dbscan_soft = DBSCAN(eps=0.5, min_samples=5)
dbscan_soft.fit(X)
soft_labels = dbscan_soft.labels_
soft_scores = dbscan_soft.core_sample_indices_


```

```python
# In[7]:


dbscan_soft = DBSCAN(eps=0.5, min_samples=5)
dbscan_soft.fit(X_pca)
soft_labelspca = dbscan_soft.labels_
soft_scores = dbscan_soft.core_sample_indices_


```

```python
# In[8]:


# Plot hard clustering results
plt.scatter(X[:, 0], X[:, 1], c=y_hard)
plt.title("Hard Clustering using DBSCAN on IRIS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()


```

```python
# In[9]:


# Plot hard clustering results with PCA
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_hardpca)
plt.title("Hard Clustering using DBSCAN and PCA on IRIS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()


```

```python
# In[10]:


# Plot soft clustering results
plt.scatter(X[:, 0], X[:, 1], c=soft_labels, cmap='viridis')
plt.title("Soft Clustering using DBSCAN on IRIS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()


```

```python
# In[11]:


# Plot soft clustering results with PCA
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=soft_labelspca, cmap='viridis')
plt.title("Soft Clustering using DBSCAN and PCA on IRIS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```
